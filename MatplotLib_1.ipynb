```
#importing the matplotlib from python. plt is used as standard in all applications thats why we are also using the same so that any user can understand it.
import matplotlib.pyplot as plt
```

```
# defining our x and y axis with random variables. 
# later on the random variables will be literally random.
x = [1,2,3,4,5,6,7,8,9]
y = [6,7,4,8,3,1,4,2,9]
```

```
#in plot enter the list you defined above and basically it takes the coordinates which we want to plot
plt.plot(x,y)
plt.show() # at the last always use plt.show to show our graph.
```

```
# defining another list to show 2 lines in our graph.
x2 = [1,2,3,4,5,6,7,8,9]
y2 = [5,6,3,7,2,1,3,1,8]

# here you have defined another coordinates but how will you know which line corresponds to which ?
# in general graphs have what is called legends for these which tells us about any such info.
# so we also need to define legend for any such graph.
```

```
plt.plot(x, y, label='First Line')
plt.plot(x2, y2, label='Second Line')
plt.legend() # necessary after using labels
plt.xlabel('X - Axis --->')
plt.ylabel('Y - Axis --->')
plt.title('Practice Graph\nTry it yourself too')

plt.show()
```
