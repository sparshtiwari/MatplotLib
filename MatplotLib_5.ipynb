```
# plotting on our own data.
```

```
import matplotlib.pyplot as plt
import numpy as np
```

```
x, y = np.loadtxt('/content/sample.txt', delimiter=',', unpack=True) # using np.loadtxt from NumPy to choose our data for plotting.
plt.plot(x,y, label='Loaded from file!')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Sample - Data')
plt.legend()
plt.show()
```
